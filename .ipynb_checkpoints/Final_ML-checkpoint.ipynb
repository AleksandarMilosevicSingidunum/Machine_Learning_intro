```python
def dot(M, v):
    u = []
    for i in range(len(M)): # cols
        temp = 0.0
        for j in range(len(M[0])): # rows
            temp = temp + M[i][j] * v[j]
        u.append(temp)
    return u 


def transpose(M):
    
    N = []
    for j in range(len(M[0])): # over rows
        U = []
        for i in range(len(M)): # over col
            U.append(M[i][j])
        N.append(U)
        
    return N


def average(a):
    return sum(a) / len(a)


def std_deviation(a):
    n = len(a) # total number of samples
    mean = sum(a) / n
    var = sum((x - mean)**2 for x in a) / n
    std_dev = var ** 0.5
    return std_dev


def square(a):
    return a ** 2
```

Feature Normalization

```python
mean = np.ones(n)
std = np.ones(m)
for i in range(n):
    mean[i] = np.mean(transpose(X)[i])
    std[i] = np.std(transpose(X)[i])
    for j in range(0,m):
        X[j][i]=(X[j][i]-mean[i])/std[i]
```

Compute Hypothesis

```python
def hypothesis(theta, X, n,m):
    # intialize a placeholder to store the calculation for every sample
    h = np.ones((m, 1))
    
    # reshape theta to (1, n+1)
    theta = theta.reshape(1, n+1)
    
    # iterate via each sample within the feature matrix
    for i in range(0,m): 
        
        # perform a forward pass
        h[i] = dot(theta, X[i])
    
    # convert the output from (m, 1) to (m,)
    h = h.reshape(m) 
    return h
```

batch gradient descent 

```python
def batchGradientDescent(theta, alpha, epochs, h, X, y, n, m):
    cost = np.ones(epochs)
    for i in range(0,epochs):
        theta[0] = theta[0] - (alpha * (2/m)) * np.sum(h - y)  # calculating bias
        for j in range(1,n+1):
            theta[j] = theta[j] - (alpha * (2/m)) * np.sum((h - y) * transpose(X)[j])  # calculating weights
        h = hypothesis(theta, X, n,m)  # update hypothesis
        cost[i] = (1/m) * 0.5 * np.sum(np.square(h - y))  # update cost function
    theta = theta.reshape(1,n+1)
    return theta, cost
```

Linear Regression

```python
def linearRegression(X, y, alpha, epochs, n, m):
    bias = np.ones((m, 1))
    X = np.concatenate((bias, X), axis=1)
    theta = np.zeros(n+1)
    h = hypothesis(theta, X,n,m)
    theta, cost = batchGradientDescent(theta, alpha, epochs, h, X, y, n, m)
    return theta, cost
```

Sigmoid Function

```python
def sigmoid(z):
    return 1. /(1+np.exp(-z))
```

Init theta and theta0

```python
def initialize_theta(n):
    theta = np.zeros([n,1])
    theta0 = 0.0
    return theta, theta0
```

Forward Propagation

```python
def forwardPropagation(theta, theta0, X_train, y_train, regularization_lambda):
    z = np.dot(np.transpose(theta),X_train)
    h = sigmoid(z)

    reg = regularization_lambda * np.sum(theta ** 2)
    cost = -y_train * np.log(h) - (1 - y_train) * np.log(1 - h)
    cost = (np.sum(cost) + reg) / X_train.shape[1]

    derivative_weight = ((np.dot(X_train, ((h - y_train).T))) / X_train.shape[1])
    derivative_bias = np.sum(h - y_train) / X_train.shape[1]

    gradients = {
        'derivative_weight': derivative_weight,
        'derivative_bias': derivative_bias
    }

    return cost, gradients
```

```python
def update_parameters(theta, theta0, X_train, y_train, alpha, epochs, regularization_lambda):
    cost_list = []
    cost_list2 = []
    index = []
    
    for i in range(epochs):
        cost, gradients = forwardPropagation(theta, theta0, X_train, y_train, regularization_lambda)
        cost_list.append(cost)
        
        theta = (theta * (1 - (alpha * regularization_lambda) / m)) - (alpha * gradients['derivative_weight'] + ((regularization_lambda) * np.sum(theta)) / m)
        theta0 = theta0 - alpha * gradients['derivative_bias']

        
        if i % 10 == 0:
            cost_list2.append(cost)
            index.append(i)
    parameters = {
        'weight': theta,
        'bias': theta0
    }
    plt.plot(index, cost_list2)
    plt.ylabel('Cost')
    plt.xlabel('Epochs')
    plt.title("Learning rate: {}".format(str(alpha)))
    

    
    return parameters, gradients, cost_list
```

```python
def regularizedLogisticRegressionModel(X_train, y_train, alpha, epochs, regularization_lambda):
    theta,theta0 = initialize_theta(X_train.shape[1])
    
    parameters, gradients, cost_list = update_parameters(theta0, theta, X_train, y_train, alpha, epochs, regularization_lambda)
    
    return parameters, gradients, cost_list
```

```python
def predict(theta, theta0, X_test):
    z = sigmoid(np.dot(theta.T, X_test))
    y_predict = np.zeros((1, X_test.shape[1]))
    
    for i in range(z.shape[1]):
        if z[0, i] >= 0.5:
            y_predict[0, i] = 1
        else:
            y_predict[0, i] = 0
    
    return y_predict
```

# Neural Network

```python
from math import exp
from random import seed, random
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
def initializeNetwork(inputNumber, hiddenNumber, outputNumber):
    network = list()
    
    hiddenLayer = [{'weights': [random() for i in range(inputNumber + 1)]} for i in range(hiddenNumber)]
    network.append(hiddenLayer)
   
    outputLayer = [{'weights': [random() for i in range(hiddenNumber + 1)]} for i in range(outputNumber)]
    network.append(outputLayer)
    
    return network
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + exp(-z))

def sigmoidDer(output):
    return output * (1.0 - output)

def activate(weights, inputs):
    z = weights[-1]
    for i in range(len(weights)-1):
        z += weights[i] * inputs[i]
    return z
```

```python
def forward_propagation(network, sample):
    inputs = sample
    for layer in network:
        new_input = []
        for neuron in layer:
            z = activate(neuron['weights'], inputs)
            neuron['output'] = sigmoid(z)
            new_input.append(neuron['output'])
        inputs = new_input
    return inputs
```

```python
def backward_propagationError(network, y):
    for i in reversed(range(len(network))):
        layer = network[i]
        errors = list()
        if i != len(network) - 1:
            for j in range(len(layer)):
                error = 0.0
                for neuron in network[i + 1]:
                    error = error+ neuron['weights'][j] * neuron['delta']
                    errors.append(error)
        else:
            for j in range(len(layer)):
                neuron = layer[j]
                error = y[j] - neuron['output']
                errors.append(error)
        for j in range(len(layer)):
            neuron = layer[j]
            neuron['delta'] = errors[j]* sigmoidDer(neuron['output']) 
```

```python
def updateWeights(network, sample, alpha):
    for i in range(len(network)):
        inputs = sample[:-1]
        if i != 0:
            inputs = [neuron['output'] for neuron in network[i-1]]
        for neuron in network[i]:
            for j in range(len(inputs)):
                neuron['weights'][j] = neuron['weights'][j] + alpha * neuron['delta'] * inputs[j]
            neuron['weights'][-1] = neuron['weights'][-1] + alpha * neuron['delta']
```

```python
def trainNetwork(network, train, alpha, epochs, outputNumber):
    costs = []  # List to store the cost values
    
    for instance in range(epochs):
        errorSum = 0
        for sample in train:
            outputs = forward_propagation(network, sample)
            y = [0 for i in range(outputNumber)]
            y[int(sample[-1])] = 1
            errorSum += sum([(y[i] - outputs[i]) ** 2 for i in range(len(y))])
            backward_propagationError(network, y)
            updateWeights(network, sample, alpha)
        
        cost = errorSum / len(train)
        costs.append(cost)  # Append the cost to the list
        
        print('Instance:', instance, 'learning rate', alpha, ', cost %.2f:' % cost)
    
    # Plot the cost over epochs
    plt.plot(range(epochs), costs)
    plt.xlabel('Epoch')
    plt.ylabel('Cost')
    plt.title('Cost over Training Epochs')
    plt.show()
```

```python
def predict(network, sample):
    outputs = forward_propagation(network, sample)
    return outputs.index(max(outputs))
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
X = (X - X.min()) / (X.max() - X.min())
#y = (y - y.min()) / (y.max() - y.min())
Xtrain, Xtest, yTrain, yTest = train_test_split(X, y, test_size=0.15, random_state=1, shuffle=True)
trainData = np.c_[Xtrain, yTrain]
testData = np.c_[Xtest, yTest]

seed(1)
trainData = np.c_[Xtrain,yTrain]
testData = np.c_[Xtest,yTest]
```

```python
inputNumber = len(trainData[0]) - 1
outputNumber = len(set([sample[-1] for sample in trainData]))
hiddenNumber = len(yTrain) // (5 * (inputNumber - outputNumber))

network = initializeNetwork(inputNumber, hiddenNumber, outputNumber)
alpha = 0.5
epochs = 100

# Train the network
trainNetwork(network, trainData, alpha, epochs, outputNumber)

# Print the network after training
for layer in network:
    print(layer)
```

```python
predictions = []
for sample in Xtest:
    predicted = predict(network, sample)
    predictions.append(predicted)

correct = 0
for i in range(len(yTest)):
    print('Predicted value:', predictions[i], 'Actual value:', yTest[i])
    if predictions[i] == yTest[i]:
        correct += 1

# Calculate the confusion matrix
cm = confusion_matrix(yTest, predictions)

# Extract the values from the confusion matrix for class 0
tn_0, fp_0, fn_0, tp_0 = cm[0, 0], cm[0, 1], cm[1, 0], cm[1, 1]

# Calculate precision, recall, and F1-score for class 0
precision_0 = tp_0 / (tp_0 + fp_0)
recall_0 = tp_0 / (tp_0 + fn_0)
f1_score_0 = 2 * (precision_0 * recall_0) / (precision_0 + recall_0)

# Extract the values from the confusion matrix for class 1
tn_1, fp_1, fn_1, tp_1 = cm[1, 1], cm[1, 0], cm[0, 1], cm[0, 0]

# Calculate precision, recall, and F1-score for class 1
precision_1 = tp_1 / (tp_1 + fp_1)
recall_1 = tp_1 / (tp_1 + fn_1)
f1_score_1 = 2 * (precision_1 * recall_1) / (precision_1 + recall_1)


# Calculate overall accuracy
accuracy = correct / float(len(yTest)) * 100.0

# Print the results
print("Accuracy: {:.2f}%".format(accuracy))
print("Class 0(Benign) - Precision: {:.2f}, Recall: {:.2f}, F1-score: {:.2f}".format(precision_0, recall_0, f1_score_0))
print("Class 1(Malignant) - Precision: {:.2f}, Recall: {:.2f}, F1-score: {:.2f}".format(precision_1, recall_1, f1_score_1))
```

```python
# Load the iris dataset
iris = load_iris()
X = iris.data[:, :3]  # Consider only the first three features for visualization

def plot_data(X, centroids=None, labels=None):
    if labels is None:
        plt.scatter(X[:, 0], X[:, 1], c='b', alpha=0.6)
    else:
        plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
    if centroids is not None:
        plt.scatter(centroids[:, 0], centroids[:, 1], c='r', marker='x', s=100)
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.title('K-means Clustering')

# Function to initialize centroids
def initialize_centroids(X, k):
    seed(1)
    indices = np.random.choice(range(X.shape[0]), size=k, replace=False)
    centroids = X[indices]
    return centroids

# Function to find the closest centroids
def find_closest_centroids(X, centroids):
    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
    closest_centroids = np.argmin(distances, axis=0)
    return closest_centroids

# Function to compute centroids means
def compute_centroids(X, closest_centroids, k):
    centroids = []
    for i in range(k):
        centroid = X[closest_centroids == i].mean(axis=0)
        centroids.append(centroid)
    return np.array(centroids)

# Function to run K-means algorithm
def run_kmeans(X, k, max_iterations=10):
    centroids = initialize_centroids(X, k)
    history = [centroids.copy()]
    for i in range(max_iterations):
        closest_centroids = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, closest_centroids, k)
        history.append(centroids.copy())
    return history

# Specify the number of clusters (k)
k = 3

# Run K-means algorithm
history = run_kmeans(X, k)
```

```python
# Plot each step of K-means
for i, centroids in enumerate(history):
    closest_centroids = find_closest_centroids(X, centroids)
    plt.figure(figsize=(8, 6))
    plot_data(X, centroids, closest_centroids)
    plt.title(f'K-means Step {i}')
    plt.show()
```
